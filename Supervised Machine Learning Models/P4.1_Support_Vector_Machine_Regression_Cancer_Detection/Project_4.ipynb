```python
import pandas as pd

Data = pd.read_csv("survey lung cancer.csv")

Data 

# data encoading 

Data["GENDER"] = Data["GENDER"].map({"M": 0, "F": 1})

Data["LUNG_CANCER"] = Data["LUNG_CANCER"].map({"YES":1,"NO":0})

Data
```

```python
Q1 = Data[["AGE","SMOKING","YELLOW_FINGERS","ANXIETY","PEER_PRESSURE","CHRONIC DISEASE","FATIGUE ","ALLERGY ","WHEEZING","ALCOHOL CONSUMING","COUGHING","SHORTNESS OF BREATH","SWALLOWING DIFFICULTY","CHEST PAIN"]].quantile(0.25)


Q3 = Data[["AGE","SMOKING","YELLOW_FINGERS","ANXIETY","PEER_PRESSURE","CHRONIC DISEASE","FATIGUE ","ALLERGY ","WHEEZING","ALCOHOL CONSUMING","COUGHING","SHORTNESS OF BREATH","SWALLOWING DIFFICULTY","CHEST PAIN"]].quantile(0.75)

IQR = Q3-Q1

IQR
```

```python

outlier = (
    (Data[["AGE","SMOKING","YELLOW_FINGERS","ANXIETY","PEER_PRESSURE","CHRONIC DISEASE","FATIGUE ","ALLERGY ","WHEEZING","ALCOHOL CONSUMING","COUGHING","SHORTNESS OF BREATH","SWALLOWING DIFFICULTY","CHEST PAIN"]] < (Q1-1.5*IQR)) 
    |
    (Data[["AGE","SMOKING","YELLOW_FINGERS","ANXIETY","PEER_PRESSURE","CHRONIC DISEASE","FATIGUE ","ALLERGY ","WHEEZING","ALCOHOL CONSUMING","COUGHING","SHORTNESS OF BREATH","SWALLOWING DIFFICULTY","CHEST PAIN"]] > (Q3+1.5*IQR)))

outlier
```

```python
Data = Data[~outlier.any(axis=1)]

Data
```

```python
from sklearn.model_selection import train_test_split

x = Data.drop("LUNG_CANCER", axis=1)
y = Data["LUNG_CANCER"]

x, y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

x_train, x_test, y_train, y_test
```

```python
from sklearn.svm import SVR

svm_regression = SVR(kernel='rbf')
svm_regression.fit(x_train,y_train)

```

```python
y_pred = svm_regression.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```
