```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv("ASPR_Treatments.csv")

data
```

```python
data.info()
```

```python
data.drop(columns = ['Last Report Date','Address 1','Provider Name','Address 2', 'City' ,'State', 'Zip', 'Public Phone', 'Latitude', 'Longitude','Geopoint', 'Grantee Code', 'Appointment URL','Home Delivery','Prescribing Services Available'], inplace=True)

data
```

```python
data.isnull().sum()
```

label = LabelEncoder()

```python
data.info()
```

```python
data_under = data[data["Is COVID-19"] == True]

true_700 = data_under.sample(n=700)

true_700.shape
```

```python
data_minor = data[data["Is COVID-19"] == False]

data_minor.shape
```

```python
sample_data = pd.concat([true_700, data_minor], axis = 0)

sample_data.head()
```

```python
# suffle the dataset

data_shuffled = sample_data.sample(frac=1, random_state=42).reset_index(drop=True)

data_shuffled.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

x = data_shuffled.drop("Is COVID-19",axis = 1)
y = data_shuffled["Is COVID-19"]

# Tran Test + Again Shuffling and sampling for safty

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y, shuffle = True)

```

```python
from sklearn.linear_model import LogisticRegression


LogReg_model_output = LogisticRegression(max_iter=1000)

LogReg_result = LogReg_model_output.fit(x_train, y_train)

LogReg_result
```

```python
LogReg_ypred = LogReg_result.predict(x_train)
```

```python
from sklearn.metrics import classification_report

LogReg_Creport = classification_report(y_train, LogReg_ypred)

print(f'C_report: {LogReg_Creport}')
```

```python
LogReg_result.score(x_train, y_train)
```

```python
LogReg_result.score(x_test, y_test)
```

```python
# SVM Model

from sklearn.svm import SVC

# Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(x_train, y_train)

```

```python
# Predictions

svm_ypred = svm_model.predict(x_test)

svm_ypred
```

```python
# Evaluate 

# Accuracy
accuracy = accuracy_score(y_test,svm_ypred)
print(f"Accuracy: {accuracy:.4f}")

# Classification report
print("Classification Report:")
print(classification_report(y_test, svm_ypred))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, svm_ypred))
```

```python
# DEcision Tree

from sklearn.tree import DecisionTreeClassifier

DT_model = DecisionTreeClassifier(max_depth=3, random_state=42)
DT_model.fit(x_train, y_train)
```

```python
# Predictions

DT_ypred = DT_model.predict(x_test)

DT_ypred

```

```python
# Evaluate 

# Accuracy
accuracy = accuracy_score(y_test,DT_ypred)
print(f"Accuracy: {accuracy:.4f}")

# Classification report
print("Classification Report:")
print(classification_report(y_test, DT_ypred))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, DT_ypred))
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier

RFSF_model = RandomForestClassifier(n_estimators=100, random_state=42)

RFSF_model

```

```python
RFSF_model.fit(x_train, y_train)

RFSF_ypred = RFSF_model.predict(x_test)

RFSF_ypred

```

```python
# Accuracy
accuracy = accuracy_score(y_test, RFSF_ypred)
print(f"Accuracy: {accuracy:.4f}")

# Classification report
print("Classification Report:")
print(classification_report(y_test, RFSF_ypred))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, RFSF_ypred))
```
